My first step is to recreate the dataframe I produced in the last assignment.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model

shroomdata = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data',sep=',', header=None, usecols=[0,3,5,9], names=["Edible","Cap Color","Odor","Gill Color"])
shroomdata.replace(to_replace={"Edible":{'p': 1, 'e': 0}}, inplace=True)
shroomdata.replace(to_replace={"Cap Color":{'n':0, 'b':1, 'c':2, 'g':3, 'r':4, 'p':5, 'u':6, 'e':7, 'w':8, 'y':9}}, inplace=True)
shroomdata.replace(to_replace={"Odor":{'a':0, 'l':1, 'c':2, 'y':3, 'f':4, 'm':5, 'n':6, 'p':7, 's':8}}, inplace=True)
shroomdata.replace(to_replace={"Gill Color":{'k':0, 'n':1, 'b':2, 'h':3, 'g':4, 'r':5, 'o':6, 'p':7, 'u':8, 'e':9, 'w':10, 'y':11}}, inplace=True)

shroomdata.head()
```

I must discover which trait, Cap Color, Odor, or Gill Color, is the best indicator to determine if a mushroom is poisonous or not.

```python
shroomdata.groupby('Edible').count()
```

The results show me that there are 4208 edible mushrooms in this database but 3916 of them are poisonous. This leaves a very high chance to encounter a poisonous mushroom.

```python
shroomdata.groupby(['Edible', 'Odor']).count()
```

The results of this sort provide me with a better look at the database. Edible mushrooms have 3 specific odors. Of those 3 only one is shared by the poisonous mushrooms. This same observation was made in my last project as well. As predicted in the last assignment, if the mushroom has any other odor besides almond or anise do not eat the mushroom.

Sniffing mushrooms can be a difficult task and some people may not have a strong sense of smell. Let's move forward with the assigment and see if we can find any other indicators that may be more obvious.

```python
shroomdata.groupby(['Edible', 'Cap Color']).count()
```

Sorting by Cap Color once again backs up my data from the last project. Any mushrooms with a Cap Color of green(4) or purple(6) are edible since no mushrooms with those colors appear in the poisonous category. I will now move on to the last category and see what my results will lead me to believe.

```python
shroomdata.groupby(['Edible', 'Gill Color']).count()
```

This dataframe shows me that mushrooms with a Gill Color of orange(6) or red(9) are going to be edible. While a Gill Color of chocolate(3) or green (5) will always be poisonous.
